```python
import pandas as pd
```

```python
# Ler o arquivo csv e converter para dataframe
df = pd.read_csv(r"data\weather_data.csv", index_col="TS")

# Vizualização do dataframe
df

# RH = humidade relativa 
# TAMB = temperatura ambiente 
# GHI = radiação
```

```python
# Visualizar principais estatísticas dos dados 
df.describe()
```

##### *Filtragem de dados de humidade relativa*

```python
# Definir limites inferior e superior para detecção de outliers
lower_bound_rh = 0
upper_bound_rh = 100

print(lower_bound_rh, upper_bound_rh)

# Criando máscara para detecção de outliers
outliers_rh = (df["RH"] < lower_bound_rh) | (df["RH"] > upper_bound_rh)
```

##### *Filtragem de dados de temperatura*

```python
# Definir limites inferior e superior para detecção de outliers
lower_bound_tamb = df["TAMB"].quantile(0.05) * 0.85
upper_bound_tamb = df["TAMB"].quantile(0.95) * 1.15

print(lower_bound_tamb, upper_bound_tamb)

# Criando máscara para detecção de outliers
outliers_tamb = (df["TAMB"] < lower_bound_tamb) | (df["TAMB"] > upper_bound_tamb)
```

```python
# Localizando todos os outliers
outliers_total = (outliers_rh | outliers_tamb)

# Filtra os dados removendo os outliers detectados
df_filtrado = df[~outliers_total]

# Vizualizando estatísticas dos dados após a filtragem
df_filtrado.describe()
```

```python
# Exportando os dados para um arquivo csv
df_filtrado.to_csv("dados_filtrados.csv")
```
